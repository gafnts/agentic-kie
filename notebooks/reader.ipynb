# Document interface

`PDFReader` is a high-level orchestrator that gives both extraction modes (single-pass and agentic) a unified interface to read a PDF.

Responsabilities: 

1. Load and hold content at both document and page granularity
2. Detect empty/broken documents
3. Fall back to Textract OCR
4. Provide and format images
5. Expose a reading interface for the agent tools 
6. Serve both extraction modes transparently 

```python
from pathlib import Path

from langchain_community.document_loaders import PyMuPDFLoader

file_path = (
    Path().cwd()
    / "notebooks"
    / "data"
    / "train"
    / "documents"
    / "00a1d238e37ac225b8045a97953e845d.pdf"
)
```

## Load a document

### Single pass: Document level

```python
full_loader = PyMuPDFLoader(file_path, mode="single")
document = full_loader.load()

document
```

```python
document[0].metadata
```

```python
document[0].page_content
```

### Agentic workflow: Page level

```python
page_loader = PyMuPDFLoader(file_path, mode="page")
documents = page_loader.load()
documents
```
